```python
"""
    simple linear regression

"""

import tensorflow as tf
fire_theft_data = [
    [6.2,29],
    [9.5,44],
    [10.5,36],
    [7.7,37],
    [8.6,53],
    [34.1,68],
    [11,75],
    [6.9,18],
    [7.3,31],
    [15.1,25],
    [29.1,34],
    [2.2,14],
    [5.7,11],
    [2,11],
    [2.5,22],
    [4,16],
    [5.4,27],
    [2.2,9],
    [7.2,29],
    [15.1,30],
    [16.5,40],
    [18.4,32],
    [36.2,41],
    [39.7,147],
    [18.5,22],
    [23.3,29],
    [12.2,46],
    [5.6,23],
    [21.8,4],
    [21.6,31],
    [9,39],
    [3.6,15],
    [5,32],
    [28.6,27],
    [17.4,32],
    [11.3,34],
    [3.4,17],
    [11.9,46],
    [10.5,42],
    [10.7,43],
    [10.8,34],
    [4.8,19]]

#
# Create model 
#

# placeholder
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# Variable
W = tf.Variable(0, dtype=tf.float32)
B = tf.Variable(0, dtype=tf.float32)

# Linear function
Y_pred = W*X + B
print(Y_pred)
# loss function
loss = tf.square(Y-Y_pred)

# Optimizer
trainer = tf.train.GradientDescentOptimizer(learning_rate = 0.001)
## Optimizerê°€ W,B
updateModel = trainer.minimize(loss)
#
#  Session run
#
with tf.Session() as sess:

    # tensorflow variable init
    sess.run(tf.global_variables_initializer())
    
    for i in range(100):
        total_loss = 0.0       
        for x, y in fire_theft_data:
            _, err = sess.run([updateModel, loss], feed_dict={X:x, Y:y})            
            total_loss += err
        #print(total_loss)                    
        train_weight = sess.run(W)
        train_bias = sess.run(B)
        print("total_loss:{}\n,train_weight:{}\n,train_bias:{}".format(total_loss,train_weight,train_bias))
            #  sess.run(...) for training
    #print out trained variables
    
```
